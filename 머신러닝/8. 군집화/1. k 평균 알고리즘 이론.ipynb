#1. 정의

> 인공지능은 크게 지도 학습과 비지도 학습으로 나뉠 수 있다. 지도학습이란 과정을 의미하는 독립변수와 결과를 의미하는 종속변수가 필요하다. 즉 종속 변수인 레이블이 존재한다면 지도학습을 통해 예측 모델을 만들 수 있다. 하지만 레이블이 없는 경우 지도학습을 사용할 수 없음으로 비지도학습을 이용하여 예측 모델을 생성해야 한다.

> K-평균은 군집화(clustering)에서 가장 일반적으로 사용되는 알고리즘이다. k-평균은 군집 중심점(centroid)이라는 특정한 임의의 지점을 선택해 해당 중심에 가장 가까운 포인트들을 선택하는 군집화 기법이다.

#3.장 단점


##3.1 장점

- 일반적인 군집화에서 가장 많이 활용되는 알고리즘
- 쉽고 간결함.

##3.2 단점

- 거리 기반 알고리즘으로 속성의 개수가 매우 많을 경우 군집화 정확도가 떨어짐
- 반복 횟수가 많을 경우 수행 시간이 매우 느려짐
- 몇 개의 군집(cluster)를 선택해야 할지 가이드하기 어려움

#4. API

```
class sklearn.cluster.KMeans(
  n_clusters=8, 
  *, 
  init='k-means++', 
  n_init=10, 
  max_iter=300, 
  tol=0.0001, 
  verbose=0, 
  random_state=None, 
  copy_x=True, 
  algorithm='auto'
)
```

> n_clusters : int, default=8
- 군집화할 크기를 지정한다. 정수형이며 기본값은 8이다.

> init: {‘k-means++’, ‘random’}, callable or array-like of shape (n_clusters, n_features), default=’k-means++’
- 기본 정보는 k-means++이며 random 옵션을 설정할 수도 있다. 

> n_init : int, default=10
- 초기 중앙점이 무작위로 선택되기 때문에 최상의 결과를 얻기 위해 몇 번 초기값을 변경하여 알고리즘을 실시할지를 정하는 횟수이다

> max_iter : int, default=300
- 중심점 이동에 대한 최대 반복 횟수로 더 이상의 이동이 없으면 군집화를 종료한다.
