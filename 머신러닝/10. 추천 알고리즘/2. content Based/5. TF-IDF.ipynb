#3. TF-IDF(Term Frequency-Inverse Document Frequency)


##3.1 이해하기
단어의 빈도수의 문서에서 나타난 단어의 빈도수의 역을 곱한 값으로 의미없는 단어들을 배제하기 위한 알고리즘이다.
###3.1.1 TF(Term Frequency)
>문서에 포함되어 있는 단어의 빈도수를 의미하는 것으로 기존에 사용했던 CountVectorizer를 의미한다.

image.png

image.png

###3.1.2 DF(Document Frequency)
>문서에 단어가 존재하는지의 유무를 판단하고 이에 대한 합을 구한다.

image.png

> '과일이'라는 단어는 문서1에서 2번 문서4에서 1번 총 3번 나타났지만 DF는 문서에 단어가 존재하는지만 판단하기 때문에 문서1에 존재하고 문서4에 존재함으로 2번 존재한다고 표현한 것이다. 이렇게 문서에 존재하는 단어의 합을 DF라고 한다.

###3.1.3 IDF(Inverse Document Frequency)
>IDF는 DF의 역을 의미하는 것이나 문서의 수가 커질수록 역수의 값이 기하급수적으로 커짐으로 다음과 같이 log를 이용하였다.

image.png

>여기서 n은 문서의 총 수가 된다.

image.png

>각 데이터를 출력하면 위와 같이 IDF 정보를 얻을 수 있다.

```python
import pandas as pd

# df = pd.DataFrame({'과일이':[2], '맛없다':[1], '맛있다':[2], '바다지':[1], '수박이':[1], '여름엔':[4]})
# df.index=['총합']

# import math
# df_tmp = df.apply(lambda x : math.log(4/(1+x)))
# df_tmp

df = pd.DataFrame({'과일이':[2,'log(4/(1+2)', 0.287682], '맛없다':[1,'log(4/(1+1)', 0.693147], '맛있다':[2,'log(4/(1+2)', 0.287682], '바다지':[1,'log(4/(1+1)', 0.693147], '수박이':[1,'log(4/(1+1)', 0.693147], '여름엔':[4,'log(4/(1+4)',-0.223144]})
df.index=['총합','IDF','log(n/(1+DF(t))']
df
```
