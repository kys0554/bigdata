#1. 유클리디안 유사도


##1.1 이해하기
>유클리디안 거리를 이용하여 유사도를 측정하는 방법으로 "비슷한 단어들은 가까운 거리에 있을 것이다"라는 가정하에 접근하는 접근 방법이다.

image.png

>위의 그림을 보면 과일과 가까운 위치의 단어인 과즙을 유사한 단어라고 판단한다는 것이다. 여기서 간단한 계산식을 본다면 중학교때 배웠던 피타고라스의 정리를 생각하면 된다.

image.png

> 피타고라스의 정리는 가로의 길이를 제곱한 것과 세로의 길이를 제곱한 값은 대각선의 제곱의 값과 같다는 것으로 a와 b를 알면 c의 길이를 알 수 있다는 것이다.

> 다시 위의 좌표를 보면 과일은 (1, 1)에 위치하고 과즙은 (4, 5)에 위치한다. 여기서 x의 길이는 4-1 = 3이고 y의 길이는 5 - 1 = 4이다. 이것을 이용하여 과일과 과즙의 거리를 구하면 3의 제곱인 9와 4의 제곱인 16을 더한 25에 루트를 씌운 5이다.

image.png

>위처럼 3개의 문서가 있다. 그리고 각 문서에 나타나는 단어들을 feature로 지정하여 행렬을 만들었다. 각 문서를 형태소 분석을 통해 단어를 추출하고 각 단어가 나타난 위치를 1로 아닌 부분을 0으로 표시한다.

image.png

> 각 feature 정보는 하나의 축을 이루게 되고 각 문서들은 각 축의 데이터를 포함하여 정보를 가지게 된다.

##1.2 문서간의 거리 계산하기

image.png

> 위의 수식을 해석하면 두 점간의 거리의 제곱에 대한 합을 루트로 나눈 값이다.

image.png

> 예를 들면 문서 1과 문서2의 유사도를 다음과 같이 구할 수 있다.
* 여름 (1-1)의 제곱 = 0
* 과일 (1-0)의 제곱 = 1
* 맛있다 (1-1)의 제곱 = 0
* 수박 (0-1)의 제곱 = 1
* 총 합 2
* 2의 루트 = 1.414
문서 1과 문서 2의 유사도는 1.414가 된다.

> 같은 방법으로 문서1과 문서3의 유사도를 다음과 같이 구할 수 있다.
* 여름 (1-1)의 제곱 = 0
* 과일 (1-0)의 제곱 = 1
* 맛있다 (1-0)의 제곱 = 1
* 수박 (0-0)의 제곱 = 0
* 바다 (0-1)의 제곱 = 1
* 총 합 3
* 2의 루트 = 1.732
문서 1과 문서 2의 유사도는 1.732가 된다.
