##1.라이브러리 등록

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

##2. 함수 만들기

```python
def getDataFrame(dataset):
  transactionEncoder = TransactionEncoder()
  transformData = transactionEncoder.fit(dataset).transform(dataset)
  return pd.DataFrame(transformData, columns=transactionEncoder.columns_)

def DeleteObject(frequent_itemsets, cnt):
  frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x : len(x))
  return frequent_itemsets[frequent_itemsets['length']>=cnt]
```

##3. 데이터 입력

```python
dataset = [ 
           ['아메리카노', '카페라떼'], 
           ['카페라떼', '아메리카노', '카푸치노'], 
           ['바닐라라떼', '아메리카노'], 
           ['녹차라떼', '카페라떼', '아메리카노'], 
           ['카페모카', '아메리카노'], 
           ['아메리카노', '카페라떼'], 
           ['초콜릿', '아메리카노'], 
           ['아메리카노'], 
           ['카페모카', '카페라떼'] 
           ]
df = getDataFrame(dataset)
```

##4. 데이터 분석
> 지지도가 너무 낮아 0.2로 조정

```python
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
print(frequent_itemsets)
```

##5. 하나의 객체 제거
> 카페라떼와 아메리카노의 연관성을 확인하는 것이므로 1개의 객체들도 나타나야 한다.

```python
removed_frequent_itemsets = DeleteObject(frequent_itemsets, 1)
print(removed_frequent_itemsets)
```

##6. 신뢰도 및 향상도 구하기
>향상도(lift)가 1 미만이므로 연관성이 없다고 볼 수 있다.

> 결과 출력을 위해 향상도 기준으로 출력을 해 보았다. 향상도가 높은 것이 좋은 것이므로 ascending을 False로 설정하였다.

```python
result = association_rules(removed_frequent_itemsets, metric="confidence", min_threshold=0.3)
result.sort_values('lift', ascending=False)
```
