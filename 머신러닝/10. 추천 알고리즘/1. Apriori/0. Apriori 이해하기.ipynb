#1. 정의
> 서로 다른 객체들간의 관계를 수치로 표현하여 연관성을 분석하는 기법

#2. 연관규칙 척도

##2.1 지지도(support)
> 전체에서 특정 객체가 발생할 빈도

image.png

>지지도는 객체(x)가 발생한 개수를 모든 객체가 발생한 수(N)으로 나누어 나타난 확률 값이다. 지지도는 전체 구매도의 경향을 파악할 수 있으며 분석 결과가 쓸만한 것인지 판단할 척도가 된다.

image.png

>위와 같은 거래 내역이 있다. 이를 이용하여 지지도를 구해보도록 하자

image.png

> 지지도가 50% 미만인 상품은 거래가 활발하지 않은 것으로 판단할 수 있음으로 그 이상인 제품에 대한 내용만 확인해 보자

image.png

>	위의 결과를 분석해 보면 과자와 라면을 함께 구매할 확률이 40%이며 라면과 달걀을 함께 구매할 확률이 60%인 것을 알 수 있다.

##2.2 신뢰도(confidence)
>두 객체 간의 연관 규칙이 유용할 가능성의 척도

image.png

>두 객체가 포함된 거래 수를  X가 포함된 거래수로 나누어 나타나는 확률 값이다. 결과가 1에 가까울수록 연관성이 높은 것이며 x를 구매할 경우 y를 구매할 확률이 높다고 생각할 수 있다.

image.png

위의 결과로 알 수 있는 것은 달걀을 살 경우 라면을 살 확률이 높다는 것을 알 수 있다.

##3. 향상도(lift)
> 두 아이템의 연관 관계가 필연적인지에 대한 척도

image.png

>x->y인 신뢰도를 y의 지지도로 나눈 결과 값이다. 실행 결과 양의 정수가 나오면 x 구매 시 y를 구매할 확률이 높은 것이며 음의 정수가 나오면 둘 중 하나만 구매할 확률이 높다. 마지막으로 결과가 1일 경우 아무런 관계가 없다고 판단할 수 있다.
		

image.png

>위의 내용으로 알 수 있는 것은 달걀을 살 경우 라면을 확률이 높으며 라면을 살 경우도 달걀을 살 확률이 높다.
		
> 최종적으로 라면과 달걀을 구매하는 지지도는 60%이며 신뢰도는 라면->달걀 75%, 달걀->라면 100%이고 향상도는 둘 다 필연적으로 구매하게 된다는 것이다. 
		
> 결국 달걀 코너 옆에 라면 코너를 두면 매출이 증대될 것이다.

* 원리가 간단하고 분석이 용이함
##3.2 단점
