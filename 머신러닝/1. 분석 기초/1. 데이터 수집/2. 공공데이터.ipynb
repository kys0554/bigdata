````python
# Python 샘플 코드 #


from urllib.request import Request, urlopen
from urllib.parse import urlencode, quote_plus

url = 'http://apis.data.go.kr/B551182/pharmacyInfoService/getParmacyBasisList'
queryParams = '?' + urlencode({ quote_plus('ServiceKey') : 'THnXNk8TrE/siPk6fKutTGL4mTHBwiytf9Zdm7Vz+i9mqeTv+/hUS6lPYRUljOI+Bgq4RZDl4LvkBFmD4xOUNg==', quote_plus('pageNo') : '1', quote_plus('numOfRows') : '10', quote_plus('sidoCd') : '110000', quote_plus('sgguCd') : '110019', quote_plus('emdongNm') : '신내동', quote_plus('yadmNm') : '온누리', quote_plus('xPos') : '127.0965441345503', quote_plus('yPos') : '37.60765568913871', quote_plus('radius') : '3000' })

request = Request(url + queryParams)
request.get_method = lambda: 'GET'
response_body = urlopen(request).read()
print(response_body)
````

````python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response_body, 'html.parser')
# 병원명
yadmnm = soup.find_all('yadmnm')
# 종별코드명
sggucdnm = soup.find_all('sggucdnm')
# 시도명
sidocdnm = soup.find_all('sidocdnm')
# 주소
addr = soup.find_all('addr')
# 전화번호
telno = soup.find_all('telno')

print('병원명 : ', yadmnm)
print('종별코드명 : ', sggucdnm)
print('시도명 : ', sidocdnm)
print('주소 : ', addr)
print('전화번호 : ')
for t in telno:
  print(t.text)
````
