#1. 앙상블
> 여러 개의 분류기(Classifier)를 생성하고 그 예측을 결합함으로써 보다 정확한 최종 예측을 도출하는 기법이다. 어려운 문제를 해결하기 위해 여러명의 전문가가 모여 상의하듯 앙상블은 단일 분류기보다 신뢰도가 높은 예측값을 얻을 수 있다.

> 앙상블 학습은 보팅(Voting), 배깅(Bagging), 부스팅(Boosting)으로 나뉘게 되며 이외에도 스태킹을 포함한 다양한앙상블 방법이 존재한다.

#2. 보팅(voting)
> 투표를 통해 여러 분류기(Classifier) 중 최적의 예측 결과를 결정하는 방식이다.
* 하드 보팅 : 가장 많은 표를 받은 클래스 선정
* 소프트 보팅 : 모든 분류기들의 클래스 라벨 확률들을 다 더한 후 평균을 취한 후 가장 높은 값의 클래스 선정

image.png

> 위의 그림처럼 보팅은 하나의 데이터셋으로 서로 다른 종류의 분류기에 테스트하여 최적의 알고리즘을 도출해 낸다.

##2.1 하드 보팅(Hard Voting)
> 예측한 결과값들 중 다수의 분류기가 결정한 예측값을 최종 보팅 결과로 선정한다.

image.png

> 위의 그림에서 보듯이 4개의 데이터로 4개의 분류기에 분석한 결과 두 번째 분류기를 제외한 모든 분류기가 1을 출력했음으로 하드 보팅의 결과는 1이 된다.

##2.2 소프트 보팅(Soft Voting)
> 각 분류기 내에서 얻은 레이블별 평균을 구한 후 예측값이 높은 평균을 보팅을 설정한다.

image.png

> class1은 0.7, 0.2, 0.8, 0.9이므로 평균 0.65를 갖는다. class2는 0.3, 0.8, 0.2, 0.1로 평균 0.35를 갖는다. class1의 예측값이 더 큼으로 예측값은 class1을 사용하게 된다.

> 일반적으로 소프트웨어 보팅의 예측 성능이 좋아 더 많이 사용된다.
