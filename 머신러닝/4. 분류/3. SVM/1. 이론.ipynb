#1. SVM이란
> SVM(Support Vector Machine)은 분류나 회귀 분석이 가능하며 hyperplane을 이용하여 분류를 하는 알고리즘이다.

> 선형 분류의 목적은 분류를 잘 할 수 있도록 마진을 넓게하는 것이다.

##2.1 장점
> * 범주나 수치 예측 문제에 사용된다.
* 오류 데이터에 영향이 없다.
* 과적합이 되는 경우가 적다.

##2.2 단점
> * 최적의 모델을 찾기 위해 커널과 모델의 다양한 테스트가 필요하다.
* 다양한 테스트를 해야 함으로 속도가 느리다.

#4. 튜닝
> SVM의 성능을 향상시키기 위해 다음과 같은 옵션들을 설정한다.

##4.1 kernel
> 커널은 다음과 같이 두 가지로 나뉘게 된다.
* linear : 선형 데이터 분석
* 가우시안 (RBF) : 비선형 데이터 분석

##4.2 Cost
> cost는 이상치 처리에 대한 기준을 두게 된다. cost의 수치가 작을 수록 이상치를 허용할 가능성이 높으며 수치가 높을 수록 이상치의 존재 가능성이 낮아 진다.

image.png

> 위의 그림에서 왼쪽 그림은 cost가 적을 경우이다. 위의 그림에서 하나의 이상치를 허용했지만 margin이 넓게 되어 모델의 성능이 높아 진다. 하지만 오른쪽의 경우 margin이 좁아 과대적합이 될 가능성이 높아 새로운 데이터가 들어오면 인식률이 떨어질 수 있다.

##4.3 gamma
> gamma는 비선형 분류인 RBF에서 사용되며 하나의 데이터 샘풀이 영향력을 행사하는 거리를 결정한다. gamma가 커질 수록 표준편차가 작아지며 gamma가 낮아질 수록 표준편차는 커진다.

> SVM에서는 Cost와 gamma를 적절히 조절하여 최적의 모델을 만드는 것이 목적이다.
