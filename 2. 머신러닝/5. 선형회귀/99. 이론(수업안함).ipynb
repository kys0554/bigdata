```python
delivery = [[100, 20], 
[150, 24],
[300, 36],
[400, 47],
[130, 22],
[240, 32],
[350, 47],
[200, 42],
[100, 21],
[110, 21],
[190, 30],
[120, 25],
[130, 18],
[270, 38],
[255, 28]]
```

```python
import pandas as pd

df_delivery = pd.DataFrame(data=delivery, columns=['distance', 'minutes'])
df_delivery
```

> 위의 데이터는 거리에 따른 배달 소요 시간을 의미한다. 거리가 늘어날 수록 시간도 점점 느는 것을 확인할 수 있다.

```python
import matplotlib.pyplot as plt

plt.scatter(df_delivery['distance'], df_delivery['minutes']) 
plt.title("Time / Distance")
plt.xlabel("Delivery Distance (meter)")
plt.ylabel("Time Consumed (minute)")
plt.axis([80, 500, 15, 60])
plt.show()
```

> 이를 간단하게 차트화 시키면 위와 같이 나타난다. 만약 거리가 500m이면 60분 정도 소요될 것이라고 예측할 수 있다.

```python
def h(x, w, b):
  return (w * x) + b

# 선형회귀 그래프 데이터 생성
x_data = range(df_delivery['distance'].values.min(), df_delivery['distance'].values.max())
y_data1 = []
y_data2 = []
y_data3 = []

for x in x_data:
  y_data1.append(h(x, 0.092, 11.33))
  y_data2.append(h(x, 0.092, 0))
  y_data3.append(h(x, 0.065, 20))
# _______________________________________________

# 그래프를 그릴 책상을 만들고
fig = plt.figure()

# 책상에 도화지 2개를 추가
ax1 = fig.add_subplot()
ax2 = fig.add_subplot()
ax3 = fig.add_subplot()
ax4 = fig.add_subplot()

# 첫 번째 도화지에 그래프 그리기
ax1 = plt.scatter(x=df_delivery['distance'], y=df_delivery['minutes'])

# 두 번째 도화지에 그래프 그리기
ax2 = plt.plot(x_data, y_data1, 'r')
ax3 = plt.plot(x_data, y_data2, dashes=[1,2], color='limegreen')
ax4 = plt.plot(x_data, y_data3, dashes=[1,2], color='violet')
```

> 선형회귀에서는 위의 각 점들과의 거리가 가장 짧게 나타나는 선을 찾는 것이 목적이다.
