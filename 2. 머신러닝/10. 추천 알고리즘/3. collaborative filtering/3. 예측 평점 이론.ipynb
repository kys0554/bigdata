##3.3 예측 평점


###3.3.1 이해하기
> 아이템 기반 추천 시스템을 만들기 위해 예측 평점을 구해보자

image.png

>위의 공식은 아이템 기반의 협업 필터링에서 개인화된 예측 평점을 구하는 공식이다.
* u : user
* i : item
* N : top N(아이템의 최근접 이웃 범위 계수)
* s : 아이템에 대한 유사도(Similarity)
* R : 사용자 평점(Rating)

> 분자는 평점 유사도 x 사용자가 작성한 평점을 의미하며 분모는 유사도의 영화별 합을 의미한다.

image.png

> 위의 그림은 분자를 표현한 것이다.사용자 평점의 행과 평점 유사도의 열을 연산하여 첫 번째 데이터를 구하게 된다. 이러한 방식으로 모든 값을 연산하여 분자를 만들 수 있다.

image.png

>위의 그림은 분모를 표현한 것이다. 평점 유사도를 영화별로 합한 결과를 분모로 사용한다.

image.png

> 이렇게 구한 두 결과를 나누면 위와 같이 예상 평점을 구할 수 있다.
