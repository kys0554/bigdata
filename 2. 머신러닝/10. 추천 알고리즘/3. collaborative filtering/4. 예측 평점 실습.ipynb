###3.3.2 예측점수 구하기

```python
print('사용자 평점\n', df_ratings_movies.values)
print('영화별 유사도\n', movies_similarity)
print('영화별 유사도\n', item_based_collab.values)
print('내적\n', df_ratings_movies.values.dot(item_based_collab.values))
```

>사용자 평점은 df_ratings_movies에 있다. 배열로 값을 추출하기 위해 values 값을 사용했다. 영화별 유사도는 movies_similarity이며 movies_similarity을 DataFrame로 변환시킨  item_based_collab.values를 이용할 수도 있다. 이 두 가지를 이용하여 간단하게 내적을 구할 수 있다.

```python
import numpy as np

print(np.abs(movies_similarity))
print(np.abs(movies_similarity).sum(axis=1))
print(np.array([np.abs(movies_similarity).sum(axis=1)]))
```

>유사도의 절대값을 표현하기 위해 abs()함수를 이용했다. 절대값의 합을 구하기 위해 sum()함수를 이용했으며 열에 따른 합을 구하기 위해 axis옵션을 추가했다. 마지막으로 이전에 구한 내적이 2차원 배월로 되어 있어 형식을 맞추기 위해 2차원 배열로 값을 출력했다.

```python
def predict_rating(ratings_arr, item_sim_arr):
  return ratings_arr.dot(item_sim_arr)/np.array([np.abs(item_sim_arr).sum(axis=1)])

ratings_pred = predict_rating(df_ratings_movies.values, movies_similarity)  
ratings_pred
```
