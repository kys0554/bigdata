```python
import pandas as pd

filepath = '/content/drive/MyDrive/Colab Notebooks/5. 컨텐츠 기반/data/tmdb_5000_movies.csv'
df = pd.read_csv(filepath)
df.head()
```

```python
columns = ['id', 'title', 'genres', 'vote_average', 'vote_count', 'popularity', 'keywords', 'overview']
df_movies = df[columns]
df_movies
```

> 결측치 확인

```python
df_movies.isnull().sum()
```

```python
df_movies[df_movies['overview'].notnull()].reset_index()
```

```python
df_movies_notnull = df_movies[df_movies['overview'].notnull()]
df_movies_notnull
```

> 이상치 확인

```python
df_movies_notnull.describe()
```

> 단어 메트릭스 생성

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

vect = TfidfVectorizer(max_df=500)
count_vect = vect.fit_transform(df_movies_notnull['overview']).toarray()
count_vect
```

```python
df_vect = pd.DataFrame(count_vect, columns=sorted(vect.vocabulary_))
df_vect
```

> 코사인 거리 계산

```python
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

cosine_similarity = cosine_similarity(df_vect, df_vect)
cosine_similarity
```

```python
overview_sim_sorted = cosine_similarity.argsort()[:,::-1]
overview_sim_sorted
```

> 데이터 입력

```python
# 영화 이름을 입력 받으면 index 정보 찾기
movie_name = input('영화 제목 : ')
recommand_movies = df_movies_notnull[df_movies_notnull['title'].str.lower().str.contains(movie_name.lower())]
recommand_movies
```

> 입력 받은 데이터를 기준으로 상위 10개 데이터 추출

```python
for movie_name in recommand_movies['title'] :
  title_movie = df_movies[df_movies['title'] == movie_name]
  recommand = overview_sim_sorted[title_movie.index.values, 1:11]
  recommand_movie = df_movies.iloc[recommand.reshape(-1)]
  print('movie : ', title_movie['title'].values[0])
  print(recommand_movie[['title', 'overview']])
  print('===================================================================================')
  print('')
```
