###3.1.4 TF-IDF
>TF-IDF는 문서에 나타난 단어의 빈도수와 IDF를 곱한 결과로 유사도를 측정하는 알고리즘이다. 단순히 빈도수만 측정하는 countvectorizer에 비해 조금 더 정확한 결과를 도출해 낼 수 있다.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer()
tfvect = vect.fit(docs2)

df_tfidv = pd.DataFrame(tfvect.transform(docs2).toarray(), columns=sorted(vect.vocabulary_))
df_tfidv.index = ['문서1', '문서2', '문서3', '문서4']
df_tfidv
```

> 위와 같이 TfidfVectorizer를 이용하여 결과를 얻을 수 있으나 실제 계산되는 내용은 우리가 손으로 계산한 결과와 약간의 차이는 있다. 우리가 알려고 했던 것은 원리 파익이기 때문에 크게 신경쓰지 않아도 된다.

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_matrix = cosine_similarity(df_tfidv, df_tfidv)
cosine_matrix
```

>코사인 유사도를 이용하여 확인해 보면 1번과 유사한 내용이 문서2, 문서 4로 나타나는 것을 알 수 있으며 중복된 데이터가 존재하도라도 유사도에는 영향을 미치지 못하는 것을 볼 수 있다.
