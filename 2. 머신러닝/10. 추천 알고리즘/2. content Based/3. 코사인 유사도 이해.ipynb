#1. 이해하기
> 코사인 유사도는 각 문서간의 각도를 기준으로 유사도를 판별하는 알고리즘으로 유사도 측정 시 많이 사용된다.

image.png

>위의 그림에서 보듯이 과일과 과실의 각도인 a와 과일과 과즙의 각도인 b를 비교하여 각도가 작은쪽이 유사도가 높다고 판단한다. 또한 '과일'
과 '과일 과일'은 같은 값으로 인식되어 각도는 0이 되게 된다.

image.png

> 코사인 유사도는 위와 같은 식으로 표현할 수 있다. A와 B의 곱에 대한 합을 A의 제곱에 대한 합의 루트와 B의 제곱에 대한 합의 루트를 곱한 값으로 나눈것이 코사인 유사도가 되는 것이다.

image.png

> 그럼위의 식을 기존으로 문서1과 문서2의 유사도를 구해 보겠다.
* 분자 : 과일이(1x0)+맛없다(0x0)+맛있다(1x1)+바다지(0x0)+수박이(0x1)+여름엔(1x1) = 2
* 첫번째 분모 : 1의 제곱(과일이) + 0의 제곱(맛없다) + 1의 제곱(맛있다) + 0의 제곱(바다지) + 0의 제곱(수박이 + 1의 제곱(여름엔) = 3의 루트
* 두번째 분모 : 0의 제곱(과일이) + 0의 제곱(맛없다) + 1의 제곱(맛있다) + 0의 제곱(바다지) + 1의 제곱(수박이 + 1의 제곱(여름엔) = 3의 루트
* 분자(2) / 분모(3) = 0.6667

>위와 같이 유사도를 구할 수 있다.

> 같은 방법으로 위의 식을 구해 보겠다.
* 분자 : 과일이(2x0)+맛없다(0x0)+맛있다(2x1)+바다지(0x0)+수박이(0x1)+여름엔(2x1) = 4
* 첫번째 분모 : 2의 제곱(과일이) + 0의 제곱(맛없다) + 2의 제곱(맛있다) + 0의 제곱(바다지) + 0의 제곱(수박이 + 2의 제곱(여름엔) = 12의 루트 = 2x3의 루트
* 두번째 분모 : 0의 제곱(과일이) + 0의 제곱(맛없다) + 1의 제곱(맛있다) + 0의 제곱(바다지) + 1의 제곱(수박이 + 1의 제곱(여름엔) = 3의 루트
* 분자(4) / 분모(6) = 0.6667

>결국 빈도수가 늘어 난다고 각도가 변하는 것은 아니라는 것을 알 수 있다.
