## 1. 라이브러리 등록

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

##2. 메소드 정의

```python
def getDataFrame(dataset):
  transactionEncoder = TransactionEncoder()
  transformMatrix = transactionEncoder.fit(dataset).transform(dataset)
  return pd.DataFrame(transformMatrix, columns=transactionEncoder.columns_)
def RemoveObject(frequent_itemsets, colName, objCnt):
  frequent_itemsets['length'] = frequent_itemsets[colName].apply(lambda x : len(x))
  return frequent_itemsets[frequent_itemsets['length']>=objCnt]
```

##3. 데이터 입력

```python
dataset = [
 ['기저귀', '버터', '맥주'],
 ['기저귀', '맥주'],
 ['기저귀', '빵'],
 ['떡', '땅콩', '사이다']
]

df = getDataFrame(dataset)
```

##4. 지지도 구하기

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

##5. 신뢰도 및 향상도 구하기
> 향상도(lift)는 1.33으로 같으나 신뢰도(confidence)가 1로 가장 높은 "맥주를 사면 기저귀를 산다"가 발생할 확률이 높은 것을 예측할 수 있다.

```python
result = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
result.sort_values(['lift','confidence', 'support'], ascending=False)
```
