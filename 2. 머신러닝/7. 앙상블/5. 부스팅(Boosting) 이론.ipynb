#1. 부스팅 이해하기
> 여러 개의 약한 학습기(weak learner)를 순차적으로 학습-예측하면서 잘못 예측한 데이터에 가중치 부여를 통해 오류를 개선해 나가면서 학습하는 방식이다.

##1.1 종류
- AdaBoost(Adaptive boosting)
- Gradient Booting Machine(GBM)
- XGBoost
- LightGBM
- CatBoost

#2. ADA Boosting
> 부스팅의 대표적인 구현은 AdaBoost(Adaptive boosting)과 그래디언트 부스트가 있다.

> 에이다 부스트는 오류 데이터에 가중치를 부여하면서 부스팅을 수행하는 대표적인 알고리즘이다.

> 피처 데이터 셋에서 +와 -를 기준의 왼쪽 오른쪽으로 나누려고 한다
1. step1은 왼쪽에 두 개의 +(플러스)를 분류 기준을 나눈 것이다. 하지만 오른쪽에 +(플러스)가 세 개로 오류가 발생된다. 
2. step2에서는 +(플러스)가 세 개를 기준으로 분류를 한다.
3. step3에서는 setp2에서 나눈 기준으로 나누었지만 -(마이너스)가 세 개 포함되어 오류가 발생된다.
4. step4에서는 -(마이너스) 세 개를 기준으로 분류한다.
5. step5에서는 step4의 기준으로 분류했지만 -(마이너스)가 한 개 포함되어 오류가 발생된다.

#3. GBM(Gradient Boost Machine)
> AdaBoost와 유사하나, 가중치 업데이트를 경사하강법(Gradient Descent)를 이용하여 최적화된 결과를 얻는 알고리즘이다. GBM은 예측 성능이 높지만 Greedy Algorithm으로 과적합이 빠르게 되고 시간이 오래 걸린다 

##3.1 하이퍼 파라미터
* loss: 경사하강법에서 사용하는 cost function을 지정하며 기본값은 deviance로 특별한 경우가 아니면 변경하지 않는다.
* learning_rate: GBM이 학습을 진행할 때마다 적용하는 학습률로써 순차적으로 오류 값을 보정해가는 데 적용하는 계수이다. 0~1사이의 값으로 기본값은 0.1이다. 너무 작은 값이면 업데이트를 너무 많이 하게 되어 꼼꼼하지만 시간이 오래 걸린다. 너무 큰 값이면 반복이 완료되었는데도 최소 오류 값을 찾지 못할 수 있지만 빠르다.
* n_estimators: weak learner의 개수(생성할 트리의 개수), 역시 많을수록 성능 향상, 시간 걸림
* subsample: weak learner(개별 트리)가 학습에 사용하는 데이터의 샘플링 비율(0~1)로써 기본값은 1이며 0.5일 경우 50%의 학습데이터를 사용한다는 것이다. 과적합을 줄이려면 1보다 작은 값을 적용해야 한다.
