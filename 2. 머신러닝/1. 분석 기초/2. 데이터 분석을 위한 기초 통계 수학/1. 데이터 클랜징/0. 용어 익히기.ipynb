#1. 대표값
>대표값이란 전체를 대표할 수 있는 값을 의미한다.
* 주문하면 얼마나 걸리나요?
* 요즘 초봉은 얼마나 하나요?


##1.1 대표값의 종류


###1.1.1 평균
> 평균은 여러 종류가 있으나 간단하게 몇가지만 알아보도록 하자
* 산술평균


image.png

> 산술평균은 모든 데이틔 합에서 데이터의 개수를 나눈 값으로 우리가 일반적으로 사용하는 평균을 의미한다.
* 가중치 평균

image.png

>가중치 평균은 각 데이터에 가중치를 곱한 값들의 합에서 가중치의 합으로 나눈 것이다.

###1.1.2 중앙값
>중앙값은 데이터를 순차적으로 나열했을 경우 중간에 위치한 값을 의미한다.

image.png

>위와 같이 5개의 데이터가 있을 경우 3번째 데이터인 7이 중앙값이 된다.

image.png

>하지만 값이 짝수라면 중간 위치에 두 개의 데이터가 존재함으로 두 값의 평균값이 중앙값이 된다. 이렇듯 위치에 따른 대표값이 중앙값이다.

###1.1.3 잘못된 평균
> 평균은 "보통 그렇더라"라고 표현할 수 있는 대표값이다.

image.png

> 위의 그림에서 평균 4m라고 해서 당당히 다이빙을 했지만 중간에 우뚝 솟은 기둥같은 것이 존재한다. 수심은 깊지만 중간에 이상한 값으로 인해 평균이 낮아진 것이지만 물 속을 보지 않으면 모를 상황인 것이다.

image.png

> 1984년 버지니아 대학에서 졸업생을 대상으로 평균 초봉을 확인할 결과 대략 6천만원의 연봉이라는 결과가 나왔다. 하지만 이해 특이하게도 NBA 센터인 랄프 샘슨(223m)의 이적료가 포함되어 있어서 많이 나오게 되었다. 대략 연봉을 계산해 보면
* 6000만원 * 100명 = 60억
* 3000만원(당시 평균 예상 임금) * 99명 = 29억 7천
* 약 30억 정도가 샘슨의 이적료로 예산된다.

>이렇듯 중간에 이상한 데이터로 인해 평균이 높아지는 상황이 발생되며 기존의 데이터들과 많은 차이가 나는 값을 이상치라고 한다.

##2.3 표분 편차의 활용
> 표분 편차의 크기로 데이터의 분포를 확인할 수 있다. 데이터의 크기가 크다면 이상치가 존재한다는 것이며 이를 이용하여 다음과 같은 분야에 활용할 수 있다.
* 음성인식 시스템 : 심한 고음이거나 저음을 배제한 음성으로 인식률 극대화
* 고객 응대 메뉴얼 : 일반 응대 메뉴얼과 극열한 고객 응대 메뉴얼 제작
