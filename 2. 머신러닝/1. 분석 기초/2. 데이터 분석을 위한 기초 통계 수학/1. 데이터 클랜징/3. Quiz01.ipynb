###힌트4

```python
# 결측 열 찾기
def getMissingList(dataFrame):
  dataSeries = dataFrame.isnull().sum()
  missingSeries = dataSeries[dataSeries>0]
  return missingSeries.index.to_list()

lst = getMissingList(df2)
for row in lst:
  print(row)
```

```python
# 대여소 번호 얻기
def getNoList(dataFrame):
  noSeries  = dataFrame['대여소번호'].drop_duplicates()
  return noSeries.values.tolist()

nolst = getNoList(df2)
nolst
```

###힌트5

```python
# 대여소번호와 column 명을 전달하면 변경
def ReplaceNaN(dataFrame, no, colName):
  mean = round(dataFrame[dataFrame['대여소번호']==no][colName].mean(),0)
  dataFrame[dataFrame['대여소번호']==no] = dataFrame[dataFrame['대여소번호']==no].replace({colName:np.nan}, {colName:mean})

  return dataFrame

df2 = ReplaceNaN(df2, 103, '반납거치대')
df2
```

```python
for no in nolst:
  for row in lst:
    df2 = ReplaceNaN(df2, no, row)

df2
```
