#1. 데이터 구분


> 데이터는 크게 정형, 비정형, 반정형 데이터로 구분한다.

##1.1 정형데이터
> 정형 데이터는 미리 정해진 형식으로 구조화된 데이터이다. 예를 들어 데이터베이스의 경우 정해진 스키마가 있고 그 스키마에 의해 데이터를 저장해야 한다.

image.png

##1.2 반정형데이터
> 기본적인 구조는 있지만 형식이 자유로운 데이터이다. 예를 들어 HTML의 경우 정해진 기본 규칙은 있지만 생략되더라도 사용할 수 있으며 별도의 규칙을 지정하여 사용할 수도 있다.

image.png

##1.3 비정형 데이터
> 정해진 규칙이 없이 자유로운 데이터이다. 예를 들어 mongoDB의 경우 데이터 저장 시 key, value 형식으로 사용됨으로 형식이 자유롭다. 만약 동영상을 저장한다면 동영상의 구조를 파악하여 저장하는 것이 아니라 동영상 자체를 저장하게 되어 구조와 관계 없이 관리할 수 있게 된다.

#2. 오픈 데이터

> 데이터 분석 시 활용되는 데이터로 라이브러리 내에서 지원하는 데이터와 사이트에서 지원해 주는 데이터가 있다.

##2.1 라이브러리 지원
1. 아이리스
> 붓꽃에 대한 데이터로 꽃잎 및 꽃받침의 너비와 길이 데이터를 가지고 있다.
2. MNIST
> 이미지 분석에서 기본적으로 활용되는 데이터로 손글씨 숫자 데이터 이다.

##2.2 사이트 지원
1. CIFAR-10
> 이미지 분석 시 활용할 수 있도록 다양한 종류의 소형 이미지 데이터를 가지고 있다. 
https://www.cs.toronto.edu/~kriz/cifar.html
2. 캐글
> 2010년 설립된 빅데이터 솔루션 경진대회 플렛픔으로 다양한 데이터를 구할 수 있다.
 https://www.cs.toronto.edu/~kriz/cifar.html
3. 공공데이터
> 국가에서 운영하는 사이트로 여러 데이터들을 얻을 수 있다. 
 https://www.data.go.kr/
